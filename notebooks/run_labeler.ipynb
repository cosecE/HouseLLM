```python
from openai import OpenAI  # changed
import pandas as pd
import json
from labeler import label_one_dialogue

client = OpenAI()  # changed
df = pd.read_csv('/content/drive/MyDrive/HouseLLM/labeler/train.csv')

results = []
for _, row in df.head(3).iterrows():
    label = label_one_dialogue(client, row['dialogue'])
    results.append({
        'encounter_id': row['encounter_id'],
        'dataset': row.get('dataset'),
        'label': label.model_dump(),
    })
    print(f"[ok] {row['encounter_id']}")

print(json.dumps(results[0], indent=2))
```

```python
output_path = '/content/drive/MyDrive/HouseLLM/labeler/labels.jsonl'

# Resume: load already-done IDs
import os, json
already_done = set()
if os.path.exists(output_path):
    with open(output_path) as f:
        for line in f:
            already_done.add(json.loads(line)['encounter_id'])

with open(output_path, 'a') as fout:
    for _, row in df.iterrows():
        eid = row['encounter_id']
        if eid in already_done:
            continue
        try:
            label = label_one_dialogue(client, row['dialogue'])
            record = {
                'encounter_id': eid,
                'dataset': row.get('dataset'),
                'dialogue': row['dialogue'],
                'label': label.model_dump(),
            }
            fout.write(json.dumps(record) + '\n')
            fout.flush()
            print(f"[ok] {eid}")
        except Exception as e:
            print(f"[fail] {eid}: {e}")
```
